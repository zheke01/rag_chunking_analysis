# Анализ стратегий чанкинга для RAG

Сравнение трёх подходов к разделению текста:
1. CharacterTextSplitter
2. RecursiveCharacterTextSplitter
3. SemanticChunker

```python
# Импорты
import sys
sys.path.insert(0, '../src')

from loaders import load_text, load_sample_texts
from chunking import CharacterSplitterStrategy, RecursiveSplitterStrategy, SemanticChunkerStrategy
from analysis import compare_strategies, calculate_metrics, visualize_comparison

import pandas as pd
import matplotlib.pyplot as plt
```

## 1. Загрузка данных

```python
# Загрузить примеры текстов
texts = load_sample_texts()
print(f"Загружено текстов: {len(texts)}")
```

## 2. Инициализация стратегий

```python
# Инициализировать стратегии
strategies = {
    'CharacterTextSplitter': CharacterSplitterStrategy(),
    'RecursiveCharacterTextSplitter': RecursiveSplitterStrategy(),
    'SemanticChunker': SemanticChunkerStrategy()
}
```

## 3. Тестирование стратегий

```python
# Выбрать текст для анализа
test_text = list(texts.values())[0]

# Сравнить стратегии
results = compare_strategies(test_text, strategies)
print(results)
```

## 4. Метрики сравнения

```python
# Вычислить метрики для каждой стратегии
metrics = {}
for strategy_name, chunks in results.items():
    metrics[strategy_name] = calculate_metrics(chunks)

# Представить в виде таблицы
df_metrics = pd.DataFrame(metrics).T
df_metrics
```

## 5. Визуализация результатов

```python
# Визуализировать сравнение
visualize_comparison(results)
```
